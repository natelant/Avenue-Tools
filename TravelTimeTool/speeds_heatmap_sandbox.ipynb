```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data - replace this with your own data
x = np.random.randint(0, 10, 100)
y = np.random.randint(0, 10, 100)
values = np.random.rand(100)

# Create a 2D histogram
heatmap, xedges, yedges = np.histogram2d(x, y, bins=(np.arange(11), np.arange(11)), weights=values)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap, annot=True, fmt=".2f", cmap="RdYlGn", xticklabels=np.arange(10), yticklabels=np.arange(10))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Heatmap')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data - replace this with your own data
x = np.random.rand(100)
y = np.random.rand(100)
values = np.random.rand(100)

# Create a contour plot
plt.figure(figsize=(8, 6))
sns.kdeplot(x=x, y=y, cmap="viridis", fill=True, thresh=0.5, levels=20, cbar=True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Map')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the CSV file
df = pd.read_csv('data/heatmap_sample/Pioneer_EB.csv')
# Convert 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Sample data - replace this with your own data
x = df['time']
y = df['offset_mi']
values = df['avg_speed']

# Create a contour plot
plt.figure(figsize=(8, 6))
sns.kdeplot(x=x, y=y, cmap="viridis", fill=True, thresh=0.5, levels=20, cbar=True)
sns.scatterplot(x=x, y=y, hue=values, palette="viridis", legend=False)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Map')
plt.xticks(rotation=45)
plt.show()
```
