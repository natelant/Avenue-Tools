```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the CSV file
df = pd.read_csv('data/heatmap_sample/Pioneer_EB.csv')
# Convert 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Sample data - replace this with your own data
x = df['time']
y = df['offset_mi']
values = df['avg_speed']

# Create a contour plot
plt.figure(figsize=(15, 15))
#sns.kdeplot(x=x, y=y, cmap="viridis", fill=True, thresh=0.5, levels=20, cbar=True)
sns.scatterplot(x=x, y=y, hue=values, palette="RdYlGn", legend=False)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Map')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define colors for congestion levels (red to green)
colors = plt.cm.RdYlGn(values / 100)  # Adjust for your actual speed range

# Plot scatterplot map with bars
plt.figure(figsize=(10, 6))
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y[i] + 1], color=colors[i], linewidth=10)

# Customize plot
plt.colorbar(plt.cm.ScalarMappable(cmap='RdYlGn'), label='Speed')
plt.xlabel('Time')
plt.ylabel('Segment of Roadway')
plt.title('Congestion Map')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot scatterplot map with bars
fig, ax = plt.subplots(figsize=(10, 6))
for i in range(len(x)):
    ax.plot([x[i], x[i]], [y[i], y[i] + 1], color=colors[i], linewidth=10)


# Create a mappable for the colorbar
sc = ax.scatter(x, y, c=values, cmap='RdYlGn')

# Customize plot
cbar = plt.colorbar(sc, ax=ax, label='Speed')

ax.set_xlabel('Time')
ax.set_ylabel('Segment of Roadway')
ax.set_title('Congestion Map')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
